### Sobre a Biblioteca

- Matplotlib é a biblioteca de visualização de dados mais popular do Python
- Ela te dá controle sobre todos aspectas da sua figura
- Ela foi projetada para te dar um sentimento semelhante aos métodos de plotagem gráfica do MatLab

### Parte 1

```python
import matplotlib.pyplot as plt
```

```python
# pra nao precisar dar plt.show()
%matplotlib inline 
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
x
```

```python
y = x * x
```

```python
y
```

```python
# Funcional
plt.plot(x, y, 'r--')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Título')
```

```python
#plt.subplot(nrows=1, ncols=2, 1)
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r--')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'g*-')
```

```python
# OO
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x, y)
axes1.set_xlabel('Eixo X')
axes1.set_title('Titulo')
axes2.plot(y, x)
```

### Parte 2

```python
fig, ax = plt.subplots() # cria a figura e o eixo
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('titulo')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
ax
```

```python
for axe in ax:
    axe.plot(x, y, 'b')
    axe.set_title('titulo')
fig
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('titulo')

ax[1].plot(x, y, 'b')
ax[1].set_title('titulo')

fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(12, 4))

axes.plot(x, y, 'r', label='hype')
axes.plot(y, x, 'g--', label='money')
axes.set_title('titulo')
# loc define que canto vai ficar a legenda, 0 é o matplot que decide qual canto
axes.legend(loc=0) 
```

```python
# salva no diretorio do notebook na extensao que quiser
fig.savefig("imagem.png")
```
