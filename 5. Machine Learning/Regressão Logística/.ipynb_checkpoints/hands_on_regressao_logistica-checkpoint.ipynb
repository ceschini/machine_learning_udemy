# Regressão Logística

Seções 4 até 4.3 do livro **Introduction to Statistical Learning**.

Aprenderemos sobre regressão logística como um método de classificação.

Alguns Exemplos de problemas de classificação:

- Filtros de e-mails (spams ou não);
- Modelos preditores de clientes inadimplentes;
- Diagnóstico de doenças.

Permite resolver problemas de classificação onde estamos tentando predizer categorias discretas.

Por convenção, classificações binárias terão duas classes: 0 e 1.

### Função Logística

```python
from IPython.display import Image
Image('../img/sigmoide.png')
```

A função logística (também conhecida por sigmoide) só retorna valores entre 0 e 1.

Isso é importante pois classificações binárias não aceitam valores fora desse intervalo.

Os valores de saída da sigmoide são interpretados como a probabilidade da classe ser 0 ou 1.

### Matriz de Confusão

Após treinarmos um modelo de regressão logística, testamos o mesmo em um conjunto de dados de teste. Porém, as métricas de avaliação de desempenho do modelo serão diferentes das métricas de Regressão Linear (MAE, MSE, RMSE).

A principal forma de avaliar o modelo é utilizando uma **matriz de confusão**.

Imagine que estamos testando uma doença:

```python
Image('../img/valores_doenca.png')
```

Exemplo: Teste para presença de doença:

- Não: Falso - 0.
- Sim: Verdadeiro - 1.

```python
Image('../img/valores_doenca_terminologia.png')
```

**Terminologia básica:**

- **True Positive (TP)**: Classifica que tem a doença e realmente tem.
- **True Negative (TN)**: Classifica que não tem a doença e realmente não tem.
- **False Positive (FP)**: Classifica que tem a doença porém não tem.
- **False Negative (FN)**: Classifica que não tem a doença porém tem.

#### Quanto em média o modelo acerta?

(TP + TN) / total = 150 / 165 = 0.91 = 91%

#### Quanto em média o modelo erra?

(FP + FN) / total = 15 / 165 = 0.09 = 9%

## Tratamento de dados

### Tratando Dados Ausentes

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Pclass', y='Age', data=train)
```

```python
# resolvendo faltantes idade
def inputar_idade(cols):
    Idade = cols[0]
    Classe = cols[1]
    
    if pd.isnull(Idade):
        if Classe == 1:
            return 37
        elif Classe == 2:
            return 29
        else:
            return 24
    else:
        return Idade
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(inputar_idade, axis=1)
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# resolvendo coluna Cabin
del train['Cabin']
# train.drop('Cabin', inplace=True)
```

```python
# resolvendo faltante embarked
train.dropna(inplace=True)
```

### Tratando Variáveis Categóricas

```python
pd.get_dummies(train['Sex'], drop_first=True)
```

```python
sex = pd.get_dummies(train['Sex'], drop_first=True)
```

```python
train['Embarked'].value_counts()
```

```python
pd.get_dummies(train['Embarked'], drop_first=True)
```

```python
embark = pd.get_dummies(train['Embarked'], drop_first=True)
```

### Tratando Variáveis Irrelevantes

```python
train
```

```python
train.drop(['Sex', 'PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
train
```

```python
# inserindo dummies
# axis = 1 nas colunas e não linhas
train = pd.concat([train, sex, embark], axis=1)
```
