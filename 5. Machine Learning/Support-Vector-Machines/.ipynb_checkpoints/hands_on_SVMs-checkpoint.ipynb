# Support Vector Machines

Seção 9 do **Introduction to Statistical Learning**.

Métodos de Machine Learning baseados em aprendizado supervisionado que analizam os dados e reconhecem padrões, sendo usados para classificação e regressão.

Dado um conjunto de dados de treino, cada um marcado com uma classe pertencendo a uma cagetoria, o SVM cria um modelo que assimila novos dados em uma das categorias, fazendo uso de um classificador não probabilistico.

Um modelo SVM é uma representação dos exemplos como pontos no espaço, mapeados de forma que as categorias são divididas com a maior distância possível entre elas.

Novos exemplos são mapeados de acordo com a zona em que caírem.

Essa divisão é feita a partir de um **hiperplano**, uma intersecção que divide o conjunto de dados pelas categorias. É possível mais de uma possibilidade de hiperplano, deseja-se escolher aquele que maximize a margem entre as classes.

```python
from IPython.display import Image
Image('../img/hiperplano.png')
```

Os pontos que tocam as linhas das margens são conhecidos como vetores de suporte (Support Vectors).

```python
Image('../img/vetores_suporte.png')
```

Podemos expandir esse conceito para dados não linearmente separáveis através do **kernel trick**.

```python
Image('../img/kernel_trick.png')
```

Se baseia em aplicar uma transformação nos dados e levá-los para um espaço dimensional superior **(2D -> 3D)** onde lá ele possa ser separado linearmente e depois os trazemos de volta aplicando a inversa.
