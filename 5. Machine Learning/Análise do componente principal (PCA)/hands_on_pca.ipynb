# Análise de Componente Principal (PCA)

Seção 10.2 do **Introduction to Statistical Learning**.

Método não supervisionado estatístico, usado para examinar relações entre um conjunto de variáveis, com o objetivo de identificar uma estrutura básica por trás das mesmas.

Também conhecido como **análise de fatores**.

Enquanto uma regressão determina a linha que melhor se ajusta ao conjunto de dados, PCA determina o melhor conjunto ortogonal de linhas para ajustar nosso modelo.

Ortogonal significa "perpendicular". Estamos procurando então por linhas perpendiculares no n-dimensional.

## PCA em Python

Iremos utilizar um dataset da biblioteca **scikit-learn** que contém 30 dimensões. Com o PCA iremos coletar os dois componentes principais dentro dessas 30 colunas e assim criarmos um gráfico que representa muito bem as duas categorias.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
cancer['target_names']
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python
scaled_data.shape
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(x_pca[:, 0], x_pca[:, 1])
plt.xlabel('Primeiro Componente principal')
plt.ylabel('Segundo Componente principal')
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=cancer['target'], cmap='plasma')
plt.xlabel('Primeiro Componente principal')
plt.ylabel('Segundo Componente principal')
```

### Entendendo o que são os componentes

```python
pca.components_.shape
```

```python
# qual a correlação de cada variavel com aquele componente
pca.components_
```

```python
df_comp = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
df_comp.head()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df_comp, cmap='plasma')
```
