# Balanço Viés-Variância

Fundamental para entender a performance de um modelo

Abordado no capítulo 2 do livro **Introduction to Statistical Learning**.

- É o que determina o ponto em que estamos apenas adicionando ruído ao modelo a medida que adicionamos complexidade ao mesmo.
- O erro nos dados de treino começa a cair a medida em que o erro nos dados de teste aumenta.
- O efeito conhecido como "overfitting" ocorre.

- Imagine que o centro do avo é o ponto que o modelo tem de prever.
- A medida que se afasta dele, as predições começam a ficar cada vez piores.

(Bias = Viés)

```python
from IPython.display import Image
Image('img/bias_variance.png') 
```

- Modelo ganha viés a medida em que a média das predições se afasta do valor correto.
- O mesmo também ganhará variância quanto mais diferentes forem as tentativas entre si.

Uma tentação comum de iniciantes é continuamente adicionar complexidade aos modelos para que os mesmos se ajustem perfeitamente aos dados de treino. O exemplo abaixo mostra um modelo simples e um complexo. No complexo o modelo não teve nenhum erro no conjunto de treino, por causa disso ele provavelmente terá mais erros que o simples no conjunto de teste.

```python
Image('img/simple_complex.png')
```

"Ao fazer isso, estamos "overfitando" nosso modelo, fazendo com que o mesmo não consiga captar o comportamento geral do conjunto de dados, apenas a se adequar ao grupo de treino. Assim, quando apresentado a novos dados de teste, o modelo terá um erro maior."

### Conclusão

É preciso tentar encontrar um equilíbrio entre a complexidade do modelo com o seu viés e variância. Adicionar complexidade só será útil até um certo ponto.
