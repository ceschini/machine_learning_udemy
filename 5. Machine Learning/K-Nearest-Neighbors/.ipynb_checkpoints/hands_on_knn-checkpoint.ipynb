# K Nearest Neighbors (KNN)

Seção 4.6 do **Introduction to Statistical Learning**.

Algoritmo de classificação de dados que opera de forma muito simples.

#### Exemplo

Imagine que temos alguns dados imaginários de alturas e pesos de cachorros e cavalos

```python
from IPython.display import Image
Image('../img/horse_dog.png')
```

 **Algoritmo de treino:**
 
 - 1) Guarde os dados.

**Algoritmo de teste/preditor:**

- 1) Calcule as distâncias do x até os demais pontos.
- 2) Organize os dados em ordem crescente de distância.
- 3) Classifique a classe de acordo com a maioria dos primeiros "k" valores.

```python
Image('../img/k_afeta.png')
```

O valor de k pode afetar as classificações, na imagem acima a estrela é da classe B se **k = 3** e é da classe A se **k = 6**.

### Pros e Cons

**Pros**:

- Muito simples.
- Processo de treino trivial.
- Funciona muito bem com um grande número de classes.
- Fácil de se adicionar mais dados.
- Poucos parâmetros (K e métrica de distância).

**Cons**:

- Elevado custo computacional para predição (pior para grandes conjunto de dados)
- Não muito bom em dados com múltiplas dimensões (muitos parâmetros)
- Parâmetros categóricas não funcionam muito bem.

## KNN com Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Classified Data', index_col=0)
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
# importante normalizar, pq as colunas podem estar em escalas diferentes
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# treinando scaler
scaler.fit(df.drop('TARGET CLASS', axis=1))
```

```python
df_normalizado = scaler.transform(df.drop('TARGET CLASS', axis=1))
```

```python
df_normalizado
```

```python
df_param = pd.DataFrame(df_normalizado, columns=df.columns[:-1])
```

```python
df_param
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_param, df['TARGET CLASS'], test_size=0.3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

### Prevendo com 1 vizinho

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

### Método do Cotovelo

```python
error_rate = []

for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_rate.append(np.mean(pred != y_test)) # salvando media de erros
```

```python
error_rate
```

```python
plt.figure(figsize=(14, 8))
plt.plot(range(1, 40), error_rate, color='blue', linestyle='dashed', marker='o')
plt.xlabel('K')
plt.ylabel('Taxa de erro')
```

```python
error_rate = pd.DataFrame(error_rate)
```

```python
error_rate.idxmin() # k = 3 foi o melhor dos 40
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
pred = knn.predict(X_test)
```

```python
print(classification_report(y_test, pred))
```
